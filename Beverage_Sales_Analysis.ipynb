```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'unit_price_(usd)':'unit_price'})
```

```python
# create a new column 'Discount_applied' based on the 'Discount' column
bins = [0, 0.05, 0.10, 0.15, 0.20]
labels = ['5%', '10%', '15%', '20%']
df['Discount_applied'] = pd.cut(
    df['Discount'],
    bins=bins,
    labels=labels,
    include_lowest=True)
```

```python
# Converting 'Order_Date' to datetime format and extracting the year into a new column 'Order_Year'
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)
df['Order_Year'] = df['Order_Date'].dt.year
```

```python
df[['Order_Date', 'Order_Year']].head(10)
```

```python
# Creating a new column Gross_sales
df['Gross_Sales'] = df['Unit_Price'] * df['Quantity']
```

```python
df[['Gross_Sales','Total_Price']].head(10)
```

```python
# Creating a new column Net_Sales 
df['Net_Sales'] = df['Gross_Sales'] * (1 - df['Discount'])
```

```python
df[['Net_Sales', 'Gross_Sales']].head(10)
```

```python
# Creating a new column Order_Size for categorizing orders based on quantity
df['Order_Size'] = np.where(df['Quantity'] <= 5, 'Small', 
                            np.where(df['Quantity'] <= 20 , 'Medium', 'Large')
                            )
                            
                        
```
